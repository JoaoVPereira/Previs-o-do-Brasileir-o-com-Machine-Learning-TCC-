# Gerando as Features


## Sequência de vitórias (Streak)

```python
brasileirao['Data'] = pd.to_datetime(brasileirao['Data'])
brasileirao['Temporada'] = brasileirao['Data'].dt.year
```

```python
#Streak
brasileirao['Home Wins'] = [1 if x == 0 else 0 for x in brasileirao['Resultado'].values]
brasileirao['Away Wins'] = [1 if x == 1 else 0 for x in brasileirao['Resultado'].values]
```

```python
def find_streak(ser):
    streak = [0]
    
    [streak.append(streak[-1]+1) if win==1 else streak.append(0) for win in ser.values]
    
    return streak[0:-1]

brasileirao['Streak Vitória Mandante'] = brasileirao.groupby(['Temporada', 'Time_Mandante'])['Home Wins'].transform(lambda x: find_streak(x))
brasileirao['Streak Derrota Mandante'] = brasileirao.groupby(['Temporada', 'Time_Mandante'])['Away Wins'].transform(lambda x: find_streak(x))
brasileirao['Streak Vitória Visitante'] = brasileirao.groupby(['Temporada', 'Time_Visitante'])['Away Wins'].transform(lambda x: find_streak(x))
brasileirao['Streak Derrota Visitante'] = brasileirao.groupby(['Temporada', 'Time_Visitante'])['Home Wins'].transform(lambda x: find_streak(x))
```

```python
brasileirao.drop(['Home Wins', 'Away Wins'], inplace=True, axis=1)
```

## Gols marcados e sofridos


```python
#Gols marcados e sofridos
def find_goals(ser):
    goals = [0]
    
    [goals.append(goals[-1]+goal) for goal in ser.values]
    
    return goals[0:-1]
```

```python
brasileirao['Goals Scored at Home'] = brasileirao.groupby(['Temporada', 'Time_Mandante'])['Mandante_Placar'].transform(lambda x: find_goals(x))
brasileirao['Goals Conceded at Home'] = brasileirao.groupby(['Temporada', 'Time_Mandante'])['Visitante_Placar'].transform(lambda x: find_goals(x))

brasileirao['Goals Scored Away'] = brasileirao.groupby(['Temporada', 'Time_Visitante'])['Visitante_Placar'].transform(lambda x: find_goals(x))
brasileirao['Goals Conceded Away'] = brasileirao.groupby(['Temporada', 'Time_Visitante'])['Mandante_Placar'].transform(lambda x: find_goals(x))
```

```python
brasileirao.tail(5)
```

## Dias desde o último jogo

```python
#Dias desde o ultimo jogo
import math
temporadas = brasileirao['Temporada'].value_counts().index.sort_values().values


brasileirao['Days Between'] = 0
brasileirao['Days Between Away'] = 0


for temporada in temporadas:
    dft = brasileirao[brasileirao['Temporada'] == temporada]
    
    for index, row in dft.iterrows():
        dfc = dft[(dft['Data'] <= row['Data']) & ((dft['Time_Mandante'] == row['Time_Mandante']) | (dft['Time_Visitante'] == row['Time_Mandante']))]
        days_bet = ((dfc['Data'] - dfc['Data'].shift()).dt.days).values[-1]
        if math.isnan(days_bet):
            brasileirao.at[index, 'Days Between'] = 5
        else:
            brasileirao.at[index, 'Days Between'] = days_bet
            
        dfc = dft[(dft['Data'] <= row['Data']) & ((dft['Time_Mandante'] == row['Time_Visitante']) | (dft['Time_Visitante'] == row['Time_Visitante']))]
        days_bet_away = ((dfc['Data'] - dfc['Data'].shift()).dt.days).values[-1]
        if math.isnan(days_bet_away):
            brasileirao.at[index, 'Days Between Away'] = 5
        else:
            brasileirao.at[index, 'Days Between Away'] = days_bet_away
```

## Confrontos diretos entre os 2 times (Histórico)

```python
#Confrontos diretos
brasileirao['Confronto Mandante'] = 0
brasileirao['Confronto Visitante'] = 0
brasileirao['Confronto Empate'] = 0

clubs = brasileirao['Time_Mandante'].value_counts().index.sort_values().values

for club1 in clubs:
    for club2 in clubs:
        dfc = brasileirao[((brasileirao['Time_Mandante'] == club1) & (brasileirao['Time_Visitante'] == club2)) | ((brasileirao['Time_Visitante'] == club1) & (brasileirao['Time_Mandante'] == club2))]
        win_home = 0
        win_away = 0
        draws = 0
        for index, row in dfc.iterrows():
            brasileirao.at[index, 'Confronto Mandante'] = win_home
            brasileirao.at[index, 'Confronto Visitante'] = win_away
            brasileirao.at[index, 'Confronto Empate'] = draws
            
            if row['Resultado'] == 1:
                win_home +=1
            elif row['Resultado'] == 2:
                win_away +=1
            else:
                draws+=1
```

# Validação Cruzada


```python
from sklearn.model_selection import cross_val_score, KFold
```

```python
# 3H40
resultados_naive = []
resultados_random_forest = []
resultados_knn = []
resultados_svm = []
resultados_rede_neural = []
#Executando os testes 30 vezes (30 é um número padrão que os cientistas)
for i in range(30):
  print(i)
  kfold = KFold(n_splits=10, shuffle=True, random_state=i)
  #Primeiro parâmetro é o valor de K(padrão 10), shuffle =  true para que o algoritmo sempre misture os resultados
  #Randonstate diferente é para o algoritmo sempre executar outro valor

  #Naive
  naive = GaussianNB()
  scores = cross_val_score(naive, x_brasileirao_treinamento, y_brasileirao_treinamento, cv = kfold)
  resultados_naive.append(scores.mean())#pegando a média dos 10 valores, então terão 30 médias

  #Random Forest
  random_forest = RandomForestClassifier(criterion = 'entropy', min_samples_leaf =10, min_samples_split = 2, n_estimators = 150, random_state = 0 )
  scores = cross_val_score(random_forest, x_brasileirao_treinamento, y_brasileirao_treinamento, cv = kfold)
  resultados_random_forest.append(scores.mean())

  #KNN
  knn = KNeighborsClassifier(n_neighbors=60, metric='minkowski', p = 20, n_jobs = 1)
  scores = cross_val_score(knn, x_brasileirao_treinamento, y_brasileirao_treinamento, cv = kfold)
  resultados_knn.append(scores.mean())

  #SVM
  svm = SVC(kernel='rbf', random_state=1, C = 1.0, probability = True, tol = 0.001 )
  scores = cross_val_score(svm, x_brasileirao_treinamento, y_brasileirao_treinamento, cv = kfold)
  resultados_svm.append(scores.mean())

  #Rede Neural
  rede_neural = MLPClassifier(activation = 'logistic', verbose=True, batch_size=56, solver = 'sgd', max_iter = 10000, tol=0.00001, hidden_layer_sizes = (22,22))
  scores = cross_val_score(rede_neural, x_brasileirao_treinamento, y_brasileirao_treinamento, cv = kfold)
  resultados_rede_neural.append(scores.mean())
```

```python
resultados = pd.DataFrame({'Naive': resultados_naive, 'Random forest': resultados_random_forest,
                           'KNN': resultados_knn,'SVM': resultados_svm,
                           'Rede neural': resultados_rede_neural})
resultados
#Criando uma tabela para exibir os resultados
```
